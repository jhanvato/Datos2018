```python
import datetime as datetime
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
```

```python
nba = pd.read_csv("nba_2013.csv", low_memory=False)
```

```python
nba.head(10)
```

```python
#Limpieza de datos

distance_columns = ['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.', 'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']
# Solo columnas numericas
nba_numeric = nba[distance_columns]

# Normalize all of the numeric columns
nba_normalized = (nba_numeric - nba_numeric.mean()) / nba_numeric.std()


nba_normalized.fillna(0, inplace=True)
```

```python
x_columns = ['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.', 'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf']
y_column = ["pts"]
```

```python
# Creo una instalncia del algoritmo que va a buscar a los 5 vecinos mas cercanos
knn = KNeighborsRegressor(n_neighbors=5)
```

```python
# Le ingreso la informacion al algoritmo
nba.fillna(0, inplace=True)
knn.fit(nba[x_columns], nba[y_column])
```

```python
# Pruebo el algoritmo con el set de entrenamiento MUY MAL, pero es en forma de ejemplo!!!
predictions = knn.predict(nba[x_columns])
```
