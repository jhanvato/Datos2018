```python
#nivel educativo
postulantes_educacion = pd.read_csv("datos/datos_navent_fiuba/fiuba_1_postulantes_educacion.csv", low_memory=False)

#fecha de nacimiento y genero
postulantes_genero_y_edad = pd.read_csv("datos/datos_navent_fiuba/fiuba_2_postulantes_genero_y_edad.csv", low_memory=False)

#vistas de avisos online y offline
vistas = pd.read_csv("datos/datos_navent_fiuba/fiuba_3_vistas.csv", low_memory=False)

#postulaciones hasta el 1ro de marzo 2018
postulaciones = pd.read_csv("datos/datos_navent_fiuba/fiuba_4_postulaciones.csv", low_memory=False)

#avisos online al 8 de marzo de 2018
avisos_online = pd.read_csv("datos/datos_navent_fiuba/fiuba_5_avisos_online.csv", low_memory=False)

#detalle de avisos vistos y postulados tanto offline como online
avisos_detalle = pd.read_csv("datos/datos_navent_fiuba/fiuba_6_avisos_detalle.csv", low_memory=False)
```

# Se realizan validaciones del nivel educativo de los postulantes.
Se filtran los niveles educativos que no estan definidos (los llamados 'Otros')
Nota: Se podria considerar 'Otros' como un estudio extra-curricular del postulante

```python
postulantes_educacion = postulantes_educacion[postulantes_educacion['nombre']!='Otro']
```

```python
# Se observa que existen postulantes con mas de un nivel educativo (valga la redundancia)
postulantes_educacion['idpostulante'].value_counts().head()
```

```python
postulantes_educacion[postulantes_educacion['idpostulante']=='Ez0LBk9'].sort_values('nombre')
```

Hasta aca se concluye que no hay problema que un postulante tenga mas de un nivel educativo ya sea en curso o no.

```python
print ( "Cantidad de vistas: " + str(len(vistas[vistas['idpostulante']=='Ez0LBk9'])) )
print ( "Cantidad de postulaciones: " + str(len(postulaciones[postulaciones['idpostulante']=='Ez0LBk9']) ) )
```

```python
postulaciones_Ez0LBk9 = postulaciones[postulaciones['idpostulante']=='Ez0LBk9'].sort_values('fechapostulacion', ascending=False)
postulaciones_Ez0LBk9
```

# Analisis de la cantidad de titulos Universitarios que tienen los postulantes  

Se procede en base a los siguientes datos filtrados

```python
postulantes_educacion[postulantes_educacion['idpostulante']=='Ez0LBk9'].sort_values('nombre')
```

```python
postulantes_educacion['cant_educacion_doble'] = 1
cantidad = postulantes_educacion.groupby(['idpostulante', 'nombre', 'estado']).agg({'cant_educacion_doble':'count'})
cantidad = cantidad.sort_values(by=['idpostulante'], ascending = False)

```

Se puede concluir que no hay postulantes con doble educacion, es decir que no hay un postulante que tenga 2 titulos univeristarios.
